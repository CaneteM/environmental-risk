```python
import math 
import numpy as np 



#Densidad del Agua

Tagua = 288.15
tagua = 15

Dagua = 1000*(1-(((tagua+288.9414)*(tagua - 3.9863)**2)/(508929.2*(tagua+68.12963))))



Taire = 300.33
# Ecuación de Antoine alternativa

Avap = 31.9407
Bvap = -2.6614e3
Cvap = -8.5763e0
Dvap = -6.7136e-10
Evap = 2.9732e-6

taire = 27.18
PM = 153.82
R2 = 8.20573645e-5




Pvap_mmhg = 10**(Avap + (Bvap/Taire) + (Cvap*np.log10(Taire)) + (Dvap*Taire) + (Evap*Taire**2))
Pvap=Pvap_mmhg/760

#Presión atmósferica 
#M es el valor de la masa molar del aire, R es la constante de los gases con Pa*m^3/(kmol*K), 
#la P se refiere a la presión sobre el nivel del mar, está cambia junto con Z0 y sus unidades son Pa.
# Tatm es la temperatura absoluta el aire en unidades de K.

P = 1
M = 28.8
g = 9.80655 
 
Z = 1207
Z0 = 0
R = 8314.462

Patmpa = P*np.exp(-((M*g)/(R*Taire))*(Z-Z0))


#Densidad del Aire

Daire = ((Patmpa*M)/(R2*Taire))/1000

Fs = Patmpa/Pvap


#Solubilidad 
#Es la cantidad máxima de un soluto que puede disolverse en
#agua pura hasta alcanzar la saturación, dada una temperatura específica.La solubilidad 
#queda en g/m^3 y las temperaturas se reportan en K. 

As = -2.15669008e+02
Bs = 1.90637080e+05
Cs = -5.54516634e+07
Ds = 5.37921624e+09


S = 10**(As + (Bs/Tagua) + (Cs/Tagua**2) + (Ds/Tagua**3))


#Constante de Henry 
#la Constante de Henry en el equilibrio es adimensional, y su valor
#depende de la temperatura de la fase gaseosa. Sin embargo, existe una forma
#dimensional H de la Constante de Henry, cuyas unidades son de la forma de
#volumen*presión/mol. Así, H es independiente de la temperatura, mientras que la
#forma adimensional Heq (la que asume un equilibrio) es válida únicamente a la
#temperatura específica del sistema.
#Constante de Henry enquivalente

#Cg = 
#CL = 

#Heq = Cg/CL


Tm = 250.15
Tb = 349.95
tb = 76.8

if Fs>=10:
    
    H = np.exp(10.6*(1-(Tb/Taire))+6.8*(1-(Tm/Taire))+0.0318*tb -5.15)                 
else:
    
    H = 10*(np.exp(10.6*(1-(Tb/Taire))+6.8*(1-(Tm/Taire))+ 0.0318*tb -5.15))

##Relación entre H equivalente y H.
# Tamb,abs es la temperatura ambiente
#absoluta (es decir, la temperatura del aire, en K), y H está en unidades de la forma de
#volumen*presión/mol, consistentes con las unidades de R. De este modo, puesto que
#las unidades típicas de H son m3atm/mol, entonces el valor del parámetro R debe ser
#8.20573645×10-5m3





Heq = H/(R2*Taire)

Kwa = 1/Heq

# Coeficiente de Partición en el equilibrio 

#Coeficiente de partición Agua-Aire, Kwa es definido como una relación entre concentración de la sustancia en el agua Cw y 
# la concentración de una sustancia en el aire Ca, se encuentran en las unidades de ppm debido a la solubilidad. 

#Cw = 
#Ca = 
#Kwa = Cw_molKg/Ca

S_gKg=S*0.001
Cw_molKg=S_gKg/PM
Cw_g_m3=Dagua*PM*Cw_molKg
Ca_molKg=Cw_molKg/Kwa
Ca_g_m3=Daire*PM*Ca_molKg





##Coeficiente de partición Carbón Orgánico-Agua. 

#Para suelos con alto contenido orgánico, el coeficiente de partición Carbón
#Orgánico-Agua, KoCw [L/kg] es definido como sigue (sólo para sustancias
#hidrosolubles):
    
    #CoC = 
    #Cw = 
Kow = 10**(2.04)
Kl = 10**(0.66*np.log10(676)-(0.0056*PM)-2.80)
    #Kow = CoC/Cw
CoC = Kow*Cw_g_m3
    
#Donde Cw es la concentración de la sustancia en el agua, y CoC es la concentración de
#la misma sustancia en el carbono orgánico presente ya sea en el suelo o en los
#sedimentos de cuerpos hidrológicos (sobre todo de cuerpos de agua eutroficados, por ejemplo). 

##Coeficiente de partición Suelo-Agua,Kd 
#El valor de Cs: concentración en el equilibrio de la sustancia en el suelo, adsorbida al suelo. 
#Cs = 
#Cw = 

#Kd = Cs/Cw

#Kd = Cs/S

#Kd = KoCw*foc+1000*((psh + pss*Heq)/(DsueloS))
#psh es porosidad del suelo humedo
#pss porosidad del suelo seco 
#DsueloS es densidad del suelo seco y Kd está en L/Kg


##Coeficiente de partición Octanol-Agua,Kow 

#Comúnmente, en la literatura se reporta el potencial negativo del coeficiente de
#partición octanol-agua (-pKow). Dicho potencial negativo es igual al logaritmo base 10
#del coeficiente dado (Yaws, 1999):

#Kow = Coct/Cw
p = 2.83
    #math.log(Kow) = -pKow
Kow  = 10**(p)
Coct_molKg = Cw_molKg*Kow
Coct_gkg = Cw_molKg*PM

#Factor de bioconcentración BCF 

BCFJo = 10**(0.85*math.log10(Kow)-0.70)
BCFMa = 10**(math.log10(Kow)-1.32)
BCF = (BCFJo*((0.95)/(0.95+0.97))+(BCFMa*(0.97)/(0.95+0.97)))


#Presion de vapor aire y humedad relativa

PsatPa = 4.59*(101325/760)*np.exp((17.27*taire)/(237.3+taire))
Psat = PsatPa/101325
Pva = Psat*(21.6/100)


#Presion de vapor de agua del medio 

Pagua = np.exp(1.18571e1 - (3.8407e3/Tagua)-(2.16961e5/Tagua**2))


#densidad de liquido sustancia

Ap = 566.07
Bp = 0.27663
fp = 0.29000
Tc = 556.35

DL = Ap*Bp**-(((((Tc - Taire)/(Tc)))**(fp)))

#Gravedad especifica 

SG = (DL/Dagua)
#Densidad del aire seco
#presion atmosferica en atm y temperatura de aire en K.  

DA= (Patmpa*28.8)/(8.206e-2*Taire)

#Humedad abs

HABS = (Pva*18.015)/(Patmpa*28.8)


#Densidad del aire Humedo

Y= (Patmpa/(8.206e-2*Taire))*((28.8*(1-HABS))+(18.015*HABS))

#Temperatura de rocio del aire
#La presion de vapor esta en atmosferas 

T_Rocio=np.linspace(-100,200,100000)
Tr= Pva-(4.596/760)*np.exp((17.27*T_Rocio)/(237.3+T_Rocio))
#plt.plot(T_Rocio,Tr)
l=pd.DataFrame([T_Rocio,Tr]).T
interval = l[l[1]<0.00001]


Resp1={'Temperatura del aire':Taire,'presion de vapor aire':Pva,'Humedad absoluta':HABS,'Densidad del aire húmedo':Y,'densidad aire seco':DA,'BCF':BCF,'BCFMa':BCFMa,'BCFJo':BCFJo,'Kow L/kg':Kow,'Coct mol/Kg':Coct_molKg,'Coct g/Kg':Coct_gkg,' Ca mol/kg':Ca_molKg,'Ca g/m3': Ca_g_m3, 'Cw mol/Kg':Cw_molKg,'Cw g/m3':Cw_g_m3}

Resp2={'Presion de vapor agua entorno':Pagua,'Gravedad especifica':SG,'Densidad de la sustancia':DL,'Solubilidad g/kg':S_gKg,'Densidad del agua':Dagua,'Coeficiente de KoW':Kow,'Densidad del aire kg/m3':Daire,'Kwa': Kwa, 'Constante de Henry': H, 'Solubilidad ppm' : S,'Factor de solubilidad': Fs, 'Constante de Henry equ': Heq, 'Presión atmosferica atm': Patmpa,'Presión Vapor atm': Pvap }

print(Kl,Resp1,Resp2)
```
